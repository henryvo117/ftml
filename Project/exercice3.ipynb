```python
import numpy as np
```

## OLS estimator simulation

```python
def OLS_estimator(X, Y):
    covariance_matrix = X.transpose().dot(X)
    inverse_covariance = np.linalg.inv(covariance_matrix)
    theta_hat = inverse_covariance.dot(X.transpose().dot(Y))
    return theta_hat

def error(Y, Y_pred):
    return sum((Y_pred - Y) ** 2)

def sigma2_calculate(Y, Y_pred, n, d):
    MSE = sum((Y_pred - Y) ** 2)
    return MSE / (n - d)
```

```python
# Size
n = 100000
d = 3

# Unknown
theta_star = np.random.normal(size=(d,1))
sigma = np.abs(np.random.normal())
eps = np.random.normal(scale=sigma, size=(n,1))


# Known
X = np.random.normal(size=(n, d))
Y = X.dot(theta_star) + eps
theta_hat = OLS_estimator(X, Y)
```

```python
Y_pred = X.dot(theta_hat)
```

```python
MSE = error(Y, Y_pred)
sigma2_hat = sigma2_calculate(Y, Y_pred, n, d)
```
