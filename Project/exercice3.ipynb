```python
import numpy as np
```

```python
# Size
n = 100
d = 10

# Unknown
theta_star = np.random.normal(size=(d,1))
sigma = np.abs(np.random.normal())
eps = np.random.normal(scale=sigma, size=(n,1))


# Known
X = np.random.normal(size=(n, d))
Y = X.dot(theta_star) + eps
```

```python
X
```

```python
Y
```

```python
XtX = X.transpose().dot(X)
XtX_inv = np.linalg.inv(XtX)
XtY = X.transpose().dot(Y)
theta_hat = XtX_inv.dot(XtY)
```

```python
Y_pred = X.dot(theta_hat)
```

```python
MSE = sum((Y_pred - Y) ** 2)
sigma2_hat = MSE / (n - d)
```
